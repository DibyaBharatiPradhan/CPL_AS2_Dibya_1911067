## Operations on complex numbers

```python
%run my_functions_library.ipynb # Running the entire library here because 
                                # importing does not work with JupyterLab

a=5
b=7
print("The 1st complex number is "+str(a)+" + "+str(b)+"i")               # printing a+bi
c=3
d=2
print("The 2nd complex number is "+str(c)+" + "+str(d)+"i")               # printing c+di

obj=MyComplex(a,b,c,d)                                                 # calling the class

p,q=obj.sum_complex(a,b,c,d)
print("Sum of "+str(a)+" + "+str(b)+"i and "+str(c)+" + "+str(d)+"i is = "+str(p)+" + "+str(q)+"i")

e,f=obj.difference_complex(a,b,c,d)
print("Difference of "+str(a)+" + "+str(b)+"i and "+str(c)+" + "+str(d)+"i is = "+str(e)+" + "+str(f)+"i")

m,n=obj.product_complex(a,b,c,d)
print("Product of "+str(a)+" + "+str(b)+"i and "+str(c)+" + "+str(d)+"i is = "+str(m)+" + "+str(n)+"i")

w,x=obj.conjugate_complex(a,b)
print("Conjugate of "+str(a)+" + ("+str(b)+"i) is = "+str(w)+" + "+str(x)+"i")

y=obj.modulus_complex(a,b)
print("Modulus of "+str(a)+" + "+str(b)+"i is = "+str(y))

r,s=obj.divide_complex(a,b,c,d)
print("Division of "+str(a)+" + "+str(b)+"i by "+str(c)+" + "+str(d)+"i gives = "+str(r)+" + "+str(s)+"i")

y=obj.phase_complex(a,b)
print("Phase angle of "+str(a)+" + "+str(b)+"i is = "+str(y)+" degrees")
```

## Avg. dist. bet^n 2 points on a st. line of N discrete points.

```python
%run my_functions_library.ipynb         # Running the entire library in one line because 
                                        # importing does not work with JupyterLab

import matplotlib.pyplot as plt
plt.figure(figsize=(9,6))

n=10                                                                           # n= number of observations plotted
index=[]
avg_dist=[]
for i in range(2,n+1):
    index.append(i)                                                  # appending the index list 
    avg_dist.append(calculate_avg_dist(i))                 # calculating average distance using i and appending to avg_dist list 

print("No. of points      Average distance")                  # to print number of points vs avg distance table
for i in range(len(index)):
    print("   "+str(index[i])+"     -------->    "+str(avg_dist[i]))

                                                                              # to label data points
for x,y in zip(index,avg_dist):
    label = "{:.2f}".format(y)
    plt.annotate(label,                                                           # this is the text
                 (x,y),                                                    # these are the coordinates to position the label
                 textcoords="offset points",                                      # how to position the text
                 xytext=(0,10),                                                   # distance from text to points (x,y)
                 ha='center')                                               # horizontal alignment can be left, right or center

                                                                            # plots number of points vs avg distance
plt.plot(index, avg_dist,'g-o')
plt.grid(color='g', ls = '-.', lw = 0.5)
plt.xlabel('No. of points')
plt.ylabel('Average distance')                                              # labeling x and y axes 
plt.title('Graph of Average distance vs No. of points')
plt.xlim(1,n+1)                                                              # setting the limit of x and y axes 
plt.ylim(0,avg_dist[-1]+1)
plt.show()
```

## Hangman Problem 

```python
%run my_functions_library.ipynb 


print("Please enter your guess in small letters since the module is case sensitive.\n")
print("Also,please enter one letter at a time")
country, capital=country_capital_guess()                                 # calling the function
print("Hint: Enter the capital of "+ country)
print("\n\n")
w=capital                                                 
w_2=list(w)                                                                       # converting into list

chance=[]                                                                 # list to store the no. of chances 
for i in range(len(w_2)):                                                          # initializing
    chance.append('_')
print(chance)                                                              

ch_left=int(len(w_2)*0.4)                                                 # to store the chances left
flag=0                                                                          # variable to declare win

i=0
while ch_left>0:
    count=0                                                                   # variable to count the number of chances left
    print("\nYou have " +str(ch_left)+ " more chances left. ")
    print()
    g=input("Enter your guess : ")

    for j in range(len(w_2)):
        if chance[j]=='_':                                                       # loop runs only for blank spaces left
            if w_2[j]==g:
                chance[j]=g
                count=1
                flag+=1
            else:
                chance[j]="_"
    print(chance)
    if count==0:
        ch_left-=1
    
                                                                 # checking the losing condition first so that the value of 
                                                                  # ch_left is not altered by the winning condition ch_left=0
    if ch_left==0:                                                                    # losing condition
        print("\nSorry! No chances left.")
        print("YOU LOST THE GAME")
        print("_________")                                                                   
        print("|       |")
        print("|      _O_")                                                          # Hangman graphic
        print("|       |")
        print("|       /\\")
        print("|____________")
        print("|____________|")

    if flag==len(chance):                                                              # winning condition
        print("\nCongratulations, you WON the game.")
        print("\U0001f929 \U0001f929 \U0001f929 ")                  # smiley emoji
        ch_left=0
    i+=1
```
